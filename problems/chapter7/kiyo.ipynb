```python
from ortoolpy import knapsack
weight = [28, 33, 10, 45, 32]
size = [2, 3, 1, 5, 4]
capacity = 7
knapsack(size, weight, capacity)
```

```python
from pulp import *
weight = [28, 33, 10, 45, 32]
size = [2, 3, 1, 5, 4]
capacity = 7
r = range(len(size))
m = LpProblem(sense=LpMaximize) 
x = [LpVariable('x%d'%i, cat='Binary') for i in r] 
m += lpDot(weight, x) 
m += lpDot(size, x) <= capacity 
m.solve()
print((value(m.objective), [i for i in r if value(x[i]) > 0.5]))
```

# 7.2

$N = \{1, \cdots, n\}$, $M = \{1, \cdots, m\}$ とする

科目 $i\in{N}$ と教室 $j\in{M}$ について、$x_{ij}$ を以下のように設定する
$$
x_{ij} =
\begin{cases}
    1 \ \ \text{if subject $i$ is open in classroom $j$} \\
    0 \ \ o.w.
\end{cases}
$$ 
n 個すべての講義をいずれかの教室で開講するため、$\forall i \in{N} \ \sum_{j=1}^m x_{ij} = 1$ が成り立つ

また、1つの教室で複数の講義を開講することはできないため、$\forall j \in{M} \ \sum_{i=1}^n x_{ij} \leq 1$ が成り立つ

各教室について、受講者数が収容人数を超える場合は1、超えない場合は0をとるダミー変数を $y_j$ とすると、これを最小化したいので、整数計画問題は以下のようになる

\begin{align*}
    \text{min} \ &\sum_{j=1}^m y_j \\
    \text{s.t.} \ \ &r_j + My_j \geq \sum_{i=1}^n d_ix_{ij} \ \ \ \text{for} \ \ j = 1, \cdots, m  \\
    & \sum_{j=1}^m x_{ij} = 1 \ \ \ \text{for} \ \ i = 1, \cdots, n\\
    & \sum_{i=1}^n x_{ij} \leq 1 \ \ \ \text{for} \ \ j = 1, \cdots, m\\
    & \ x_{ij}\in{\{0, 1\}} \ \ \ \text{for} \ \ i=1, \cdots, n, \ j = 1, \cdots, m\\
    & \ y_{j}\in{\{0, 1\}} \ \ \ \text{for} \ \ j = 1, \cdots, m
\end{align*}

# 7.3

$m$ 個のデータ点 $s_1, \cdots, s_m \in {\mathbb{R}^d}$ について、$k$ 個のクラスター $C_1, \cdots, C_k$ に分ける非階層的クラスタリングを考える。

まず、$V = \{s_1, \cdots, s_m\}$ について、$k$ 個のクラスター中心を選ぶので、選ばれた $k$ 個のデータ点を $ U = \{s^*_1, \cdots, s^*_k\}$ とする。

次に、$s_l\in{V}$ が $s^*_i\in{U}\subset{V}$ を中心とするクラスターに含まれる場合に1をとるダミー変数を $x_{li}$ とすると、$\forall l\in{\{1, \cdots, m\}} \ \ \sum_{i=1}^k x_{li} = 1$

したがって、整数制約付き2次錐計画問題は以下のようになる
\begin{align*}
    \min_{U, \ x_{li}, \ z} \ &z \\
    \text{s.t.} \ \ &z + M(1-x_{li}) \geq \| s_l - s^*_i \|_2 \ \ \ \text{for} \ \ l = 1, \cdots, m, \ \ i = 1, \cdots, k  \\
    & \sum_{i=1}^k x_{li} = 1 \ \ \ \text{for} \ \ l = 1, \cdots, m\\
    & \ x_{li}\in{\{0, 1\}} \ \ \ \text{for} \ \ l = 1, \cdots, m, \ \ i = 1, \cdots, k
\end{align*}


# 7.4

$s_l \in{\{s_1, \cdots, s_m\}}$ がクラスター $C_i$ に含まれる場合に1をとるダミー変数を $x_{li}$ とすると、$\forall l\in{\{1, \cdots, m\}} \ \ \sum_{i=1}^k x_{li} = 1$

よって、整数計画問題は以下のようになる
\begin{align*}
    \min_{x_{li}, \ z_l} \ & \sum_{l=1}^m z_l \\
    \text{s.t.} \ \ &z_l + M(1-x_{li}) \geq \| s_l - \mu_i \|_2^2 \ \ \ \text{for} \ \ l = 1, \cdots, m, \ \ i = 1, \cdots, k  \\
    & \sum_{i=1}^k x_{li} = 1 \ \ \ \text{for} \ \ l = 1, \cdots, m\\
    & \ x_{li}\in{\{0, 1\}} \ \ \ \text{for} \ \ l = 1, \cdots, m, \ \ i = 1, \cdots, k
\end{align*}

したがって、整数制約付き2次錐計画問題は以下のようになる
\begin{align*}
    \min_{x_{li}, \ z} \ & \sum_{l=1}^m z_l \\
    \text{s.t.} \ \ &z_l + M(1-x_{li}) + 1 \geq \left|\left| \begin{bmatrix} M(1-x_{li}) - 1 \\ 2( s_l - \mu_i ) \end{bmatrix} \right|\right|_2 \ \ \ \text{for} \ \ l = 1, \cdots, m, \ \ i = 1, \cdots, k  \\
    & \sum_{i=1}^k x_{li} = 1 \ \ \ \text{for} \ \ l = 1, \cdots, m\\
    & \ x_{li}\in{\{0, 1\}} \ \ \ \text{for} \ \ l = 1, \cdots, m, \ \ i = 1, \cdots, k
\end{align*}
