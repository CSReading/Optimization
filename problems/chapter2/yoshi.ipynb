### 2.4
#### 問
- 以下のの線形計画問題において, $z$を基底変数とするこの問題の実行可能基底解はどのようなものか.また, この問題の最適解と問題2.15実行可能基底解にはどのような関係があるか.
$$
\begin{equation*}
\begin{aligned}
\min \quad & 1^T z \\
\textrm{s.t.} \quad & Ax + z = b \\
    & x \geq 0, z \geq 0
\end{aligned}
\end{equation*}
$$
- 問題2.15
$$
\begin{equation*}
\begin{aligned}
\min \quad & c^T x \\
\textrm{s.t.} \quad & Ax = b \\
    & x \geq 0
\end{aligned}
\end{equation*}
$$

#### 答
- $z$を基底変数と置いた場合, 非基底変数$x$は$x = 0$. よって, $z = b$
- 仮に最適値が$0$となる場合, $Ax = b$が成立するため, 問の問題の最適解の$x$は問題2.15の実行可能解となる.（これは常に基底になるの？解き方に依るのでは？）
- 最適値が正の値となった時は, 問題2.15は実行可能解を持たない.

### 2.5
#### 問
- 次の最適化問題を線形計画問題か凸２次計画問題に帰着せよ
- (1) チェビシェフ近似問題
$$
\begin{equation*}
\begin{aligned}
\min \quad & ||Ax - b||_\infty 
\end{aligned}
\end{equation*}
$$
- (2) l_1ノルム正則化つきチェビシェフ近似問題
$$
\begin{equation*}
\begin{aligned}
\min \quad & ||Ax - b||_\infty + \gamma ||x||_1
\end{aligned}
\end{equation*}
$$
- (3) ティコノフ正則化付きチェビシェフ近似問題
$$
\begin{equation*}
\begin{aligned}
\min \quad & ||Ax - b||_\infty + \gamma ||x||_2^2
\end{aligned}
\end{equation*}
$$
- (4) エラスティックネット付き最小２乗法
$$
\begin{equation*}
\begin{aligned}
\min \quad & ||Ax - b||_2^2 + \gamma ||x||_1 + \rho ||x||_2^2
\end{aligned}
\end{equation*}
$$

#### 答
- (1) チェビシェフ近似問題
$$
\begin{equation*}
\begin{aligned}
\min \quad & y \\
\textrm{s.t.} \quad & Ax - b \leq y*1 \\
    & - Ax + b \leq y*1
\end{aligned}
\end{equation*}
$$

- (2)  l_1ノルム正則化つきチェビシェフ近似問題
$$
\begin{equation*}
\begin{aligned}
\min \quad & y + \gamma 1^T z \\
\textrm{s.t.} \quad & Ax - b \leq y*1 \\
    & - Ax + b \leq y*1 \\
    & x \leq z \\
    & -x \leq z
\end{aligned}
\end{equation*}
$$

- (3) ティコノフ正則化付きチェビシェフ近似問題
$$
\begin{equation*}
\begin{aligned}
\min \quad & \frac{1}{2} y + \gamma x^Tx\\
\textrm{s.t.} \quad & Ax - b \leq y*1 \\
    & - Ax + b \leq y*1 \\
\end{aligned}
\end{equation*}
$$

- (4) エラスティックネット付き最小２乗法
$$
\begin{equation*}
\begin{aligned}
\min \quad & x^T(A^TA)x - 2b^TAx + \gamma x^Tx + \rho 1^Ty \\
\textrm{s.t.} \quad & x \leq y \\
    & -x \leq y
\end{aligned}
\end{equation*}
$$
